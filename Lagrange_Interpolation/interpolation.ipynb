```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
```

```python
x=np.linspace(2,10,10)
y=np.sin(x)
plt.scatter(x,y)
```

```python
def inter(t):
    n=x.shape[0]
    accum=0
    for j in range(0,n):
        prod=1
        for k in range(0,n):
            if k != j:
                prod *= (t-x[k])/(x[j]-x[k])
        accum+=y[j]*prod
    return accum
```

```python
inter_x=np.linspace(2,10,1000)
inter_y=inter(inter_x)
```

```python
plt.scatter(inter_x, inter_y, c='red', alpha=0.5)
plt.scatter(x,y, c='blue',alpha=0.5)
```

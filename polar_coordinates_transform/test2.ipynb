```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
```

```python
r = lambda t : t
phi = lambda t : t

r_hat=lambda t: r(t)*np.array([np.sin(phi(t)),np.cos(phi(t))])
phi_hat =lambda t: r(t)*np.array([np.cos(phi(t)),-np.sin(phi(t))])

PI=3.14

t=PI/4

J=r(t)*np.array([[np.sin(phi(t)),np.cos(phi(t))],[np.cos(phi(t)),-np.sin(phi(t))]])

p_x, p_y = 1,2
p = np.array([p_x,p_y])

```

```python
fig, ax = plt.subplots()

SCALE=5
v1=ax.quiver(0, 0, (J@p)[0], (J@p)[1], color='r', scale=SCALE, alpha=0.5)

N=1
ax.set_xlim([-N, N])
ax.set_ylim([-N, N])

plt.grid()
plt.show()
```

```python
time=np.linspace(1,10,10)

def update(frame):
    # for each frame, update the data stored on each artist.
    t = time[:frame]
    
    p=np.array([r_hat(t),phi_hat(t)])
    
    # update the scatter plot:
    data = np.stack([p[0], [[0]]]).T
    ax.set_offsets(data)
    # update the line plot:
    v1.U=p[0]
    v1.V=p[1]
    return (ax, v1)

ani = animation.FuncAnimation(fig=fig, func=update, frames=40, interval=30)
plt.show()
```
